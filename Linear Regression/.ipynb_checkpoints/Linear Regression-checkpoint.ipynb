```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X = pd.read_csv("./Training Data/Linear_X_Train.csv").values
y = pd.read_csv("./Training Data/Linear_Y_Train.csv").values
```

```python
plt.figure(figsize=(8,5))
plt.scatter(X,y)
```

```python
def hypothesis(x,theta):
    return theta[0]+ theta[1]*x
```

```python
def error(X,Y,theta):
    m = X.shape[0]
    error = 0 
    for i in range(m):
        hx = hypothesis(X[i],theta)
        error += (hx-Y[i])**2
    
    return error
```

```python
def gradient(X,Y,theta):
    m = X.shape[0]
    grad = np.zeros(2)
    
    for i in range(m):
        hx = hypothesis(X[i],theta)
        grad[0] += (hx -Y[i])
        grad[1] += (hx-Y[i])*X[i]
    
    return grad
```

```python
def gradientDescent(X,Y,learning_rate = 0.0005):
    
    theta = np.array([-1.0,1.4])
    
    theta_list = []
    error_list = []
    
    i=0
    max_iter = 50
    
    while(i<max_iter):
        grad = gradient(X,Y,theta)
        e = error(X,Y,theta)
        
        error_list.append(e)
        theta_list.append((theta[0],theta[1]))
        
        
        theta[0] = theta[0] - learning_rate*grad[0]
        theta[1] = theta[1] - learning_rate*grad[1]
        
        
        i+=1
        
        
    
    return theta, theta_list, error_list
```

```python
theta, theta_list,error_list = gradientDescent(X,y)
```

```python
plt.plot(error_list)
```
