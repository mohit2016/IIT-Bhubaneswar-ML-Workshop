```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('seaborn')
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
plt.plot(x, label= "linear")
plt.plot(y, label= "quad")
plt.legend()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, c="red")

plt.subplot(1,2,2)
plt.plot(y, x)


plt.show()
```

**This is called functional method of creating graphs**

```python
plt.figure(figsize=(10,6))

plt.plot(x,y)
plt.plot(y,x, linestyle=":", marker="o")
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
```

### Scatter Plot

```python
X = np.random.randn(200,2)
Y = np.random.randn(200,2)
Z = X*1 + 10
```

```python
plt.scatter(X[:,0],X[:,1])
plt.scatter(Y[:,0],Y[:,1], c="red")
plt.scatter(Z[:,0],Z[:,1], c="yellow")
```

### Bar Plot

```python
x = np.array([1,2,3])
```

```python
plt.bar(x, [10,50,30], width=0.3, label= "Last Year")
plt.bar(x+0.3, [30,20,70], width=0.3, label="This Year")
plt.legend()
plt.show()
```
